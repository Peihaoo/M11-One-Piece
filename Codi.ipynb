```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.set(style="darkgrid")
```

```python
df = pd.read_csv('Pokemon.csv')
```

```python
df.head()
```

```python
len(df)
```

```python
df.dtypes
```

```python
df.drop('Total', inplace=True, axis=1)
```

```python
df.columns = ['#', 'Nom', 'Tipus1', 'Tipus2', 'HP', 'Atac', 'Defensa', 'SPA', 'SPD', 'Velocitat', 'Gen', 'Legendary']
```

```python
df.head()
```

```python
mitjanes_per_gen = df.groupby('Gen')[['HP', 'Atac', 'Defensa', 'SPA', 'SPD', 'Velocitat']].mean()

plt.figure(figsize=(12, 7))

for columna in mitjanes_per_gen.columns:
    plt.plot(mitjanes_per_gen.index, mitjanes_per_gen[columna], marker='o', label=columna)

plt.title('Media de Estadísticas por Generación')
plt.xlabel('Generació')
plt.ylabel('Mitjana')
plt.xticks(mitjanes_per_gen.index)  # Asegura que las generaciones estén en el eje x
plt.grid(True)

# Añadir la leyenda
plt.legend(loc='upper left')

plt.tight_layout()
```

```python
plt.show()
```
